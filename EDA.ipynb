```python
import pandas as pd

calorie_dataset_path = 'data/calorie_dataset.csv'

try:
    calorie_df = pd.read_csv(calorie_dataset_path)
    read_success = True
except Exception as e:
    read_success = False
    error_message = str(e)

read_success, calorie_df.head() if read_success else error_message
```
